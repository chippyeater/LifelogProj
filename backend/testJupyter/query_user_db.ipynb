```python
import os, sqlite3

db_path = os.getenv("PIPELINE_DB_PATH", "lifelog.db")
user_id = "003"

conn = sqlite3.connect(db_path)
query = """
SELECT * FROM user_videos
WHERE user_id = ?
ORDER BY updated_at DESC
"""
cur = conn.execute(query, (user_id,))
cols = [d[0] for d in cur.description]
rows = cur.fetchall()
conn.close()
print(cols)
for r in rows:
    print(r)
```
